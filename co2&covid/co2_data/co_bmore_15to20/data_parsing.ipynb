```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hold = [0, 0, 0, 0, 0, 0] 

#load data

dataframes = ['15', '16', '17', '18', '19', '20']
names = hold.copy()
placehold = names.copy()

for x in range(6):
    placehold[x] = '20{}_CO_DATA.csv'.format(dataframes[x])
    names[x] = pd.read_csv('co_bmore_15to20\\' + placehold[x])
    names[x].dropna(inplace = True)
    
#unpack names

co15, co16, co17, co18, co19, co20 = names
```

```python
#extract aqi

aqiholder = names.copy()
for x in range(6):
    aqiholder[x] = names[x]['DAILY_AQI_VALUE']

#name aqi

aqi_15, aqi_16, aqi_17, aqi_18, aqi_19, aqi_20 = aqiholder

#extract date

dateholder = names.copy()
for x in range(6):
    dateholder[x] = names[x]['Date']
    
#name dates

date_15, date_16, date_17, date_18, date_19, date_20 = dateholder 
```

```python
#date reformat - find change in month and record the index where changes, use index to mark month

def month_index(x):
    monthHolder = []
    for i, j in enumerate(x):
        if i == 0:
            monthHolder.append(i)
        else:
            if j[0:2] != x[i-1][0:2]:
                monthHolder.append(i)
    return monthHolder
            
```

```python
plt.figure(figsize=(60,20))
plt.plot(date_15, aqi_15)
plt.title('2015 AQI Data', fontsize=70)

# x axis
plt.xlabel('Date', fontsize=60)
plt.xticks(month_index(date_15))
plt.xticks(fontsize=30)

# y axis
plt.ylabel('Air Quality Index', fontsize=60)
plt.yticks(fontsize = 30)


plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(date_20, aqi_20, label='2020')
plt.title('2020 AQI Data vs 2015 AQI Data', fontsize=20)

# x axis
plt.xlabel('Date', fontsize=15)
plt.xticks(month_index(date_20))
plt.xticks(fontsize=10)

# y axis
plt.ylabel('Air Quality Index', fontsize=15)
plt.yticks(fontsize = 10)
plt.plot(aqi_15[0:88], 'r', label='2015')


plt.grid(linewidth=1)
plt.legend()
plt.show()
```
