```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import bokeh.io
import bokeh.plotting

bokeh.io.output_notebook()
```

```python
cov = pd.read_csv('MDCOVID19_CasesByCounty.csv', na_values='0')
cov
```

```python
plt.plot(cov['OBJECTID'], cov['Howard'], 'r', label = 'Howard County')

plt.ylabel('# of confirmed cases')

plt.title('Covid confirmed cases')
plt.legend()
plt.show
```

Bokeh

```python
p = bokeh.plotting.figure(
    frame_width=400,
    frame_height=800,
    x_axis_label='date',
    y_axis_label='# of confirmed cases'
)

p.circle(
    source=cov,
    x = 'OBJECTID',
    y = 'Howard'
);

bokeh.io.show(p)
```
